### Подключение библиотек

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
```

### Загрузка датасета

```python
df = pd.read_csv("UniversalBank.csv")
df.head(5)
```

### Удаление лишних столбцов

```python
df = df.drop(columns=["Row", "ZIP Code"])
```

### Разделение наи признаки и целевую переменную

```python
X = df.drop(columns=["Personal Loan"])
y = df["Personal Loan"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=31, stratify=y)
```

### Построение дерева решений

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=12, criterion="entropy")
dt.fit(X_train, y_train)
```

```python
plt.figure(figsize=(12,8))
tree.plot_tree(dt, filled=True, class_names=["Not Accepted", "Accepted"], feature_names=list(X_train.columns))
plt.show()
```
